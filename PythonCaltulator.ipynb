```
def calculator():
  choose_action = input("""Please Select : 
                        1.Addition
                        2.Subtraction
                        3.Multiplication
                        4.Division
                        5.Exponential
                        
    (Write the number in the box :)\n""")
  #Addition
  if choose_action == '1' :
    num1 = int(input("Enter 1st number :"))
    num2 = int(input("Enter 2nd number :"))
    print(num1+num2)
  #Subtraction
  elif choose_action == '2' :
    num1 = int(input("Enter 1st number :"))
    num2 = int(input("Enter 2nd number :"))
    print(num1-num2)
  #Multiplication
  elif choose_action == '3' :
    num1 = int(input("Enter 1st number :"))
    num2 = int(input("Enter 2nd number :"))
    print(num1*num2)
  #Division
  elif choose_action == '4' :
    num1 = int(input("Enter 1st number :"))
    num2 = int(input("Enter 2nd number :"))
    print(num1/num2)
  #Exponential
  elif choose_action == '5' :
    num1 = int(input("Enter 1st number :"))
    num2 = int(input("Enter 2nd number :"))
    print(num1**num2)
  
 
  
calculator()
```
